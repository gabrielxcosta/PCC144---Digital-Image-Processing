# Questões sobre Imagens

````python
def show(img_list, caption_list, Axis=False):
    n = len(img_list)
    f, ax = plt.subplots(1,n, figsize=(10+5*(n-1),10))
    for i in range(n):
        ax[i].imshow(img_list[i], cmap='gray')
        ax[i].set_title(caption_list[i])
        if not Axis:
            ax[i].set_axis_off()
````

<p>13) Trocar entre si os canais vermelho e azul. Ou seja, a imagem R,G,B vai ficar B,G,R</p>

````python
img = data.chelsea()
R,G,B = img[:, :, 0], img[:, :, 1], img[:, :, 2]
nimg = img.copy()
nimg[:, :, 0], nimg[:, :, 2] = B, R
show([img, nimg], ['Original', 'Trocar'])
````

<p> 14) Rotacionar uma imagem em 90 graus sem utilizar funções que realizem a rotação de forma explicita</p>

````python
def rotate90(img):
  transposed = np.transpose(img, (1, 0, 2)) # inverte as dimensões da imagem, troca as linhas pelas colunas
  rotated = np.flip(transposed, axis=1) # inverte as colunas da imagem transposta, completando a rotação de 90 graus
  return rotated
````

````python
img = data.chelsea()
nimg = rotate90(img)
show([img, nimg], ['0 graus', '90 graus'])
````

````python
cat = data.chelsea()

x, y, side_length = 10, 10, 100 # dims do quadrado
cat_with_square = cat.copy() # não irei alterar a imagem original mas sim uma cópia dela
cat_with_square[y:y+side_length, x:x+side_length] = [255, 0, 0] # colocar o quadrado na posição / intensidade RGB [R, G, B], tá num intervalo [0, 255]

plt.imshow(cat_with_square)
````
